# Example 1

## $$f(x) = x^2 + x + 1$$

```python
def f(x) :
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
# Plot
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

# Slope & Derivatives

Challenge: Create a Python function for the derivative of f(x) called df(x)

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot plot function and derivative side by side

plt.figure(figsize=[15, 5])

# 1 Chart : Cost Function

plt.subplot(1, 2, 1)

plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1), color = 'blue', linewidth = 3)

# 2 Chart : Derivate
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize = 17)
plt.xlabel('X', fontsize=16)
plt.ylabel('d(x)', fontsize=16)
plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 3)
plt.show()
```
